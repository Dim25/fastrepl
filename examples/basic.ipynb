```python
import fastrepl.repl as fastrepl
```

```python
def llm_chain1(v1: str):
    v2 = fastrepl.Updatable(key="key1", value="value1").value
    return f"({v1}, {v2})"


def llm_chain2(v1: str):
    v2 = fastrepl.Updatable(key="key2", value="value2").value
    return f"({v1}, {v2})"


def llm_chain(input: str):
    v1, v2 = llm_chain1(input), llm_chain2(input)
    return f"({v1}, {v2})"
```

```python
with fastrepl.REPL():
    print(llm_chain("hello"))
```

```python
fastrepl.graph(level=2)
```

```python
fastrepl.update([("key1", "value3"), ("key2", "value4")])
```

```python
with fastrepl.REPL():
    print(llm_chain("hello"))
```
