```python
from fastrepl.polish import Updatable


def llm_chain1(input: str):
    Updatable(key="key1", value=input)
    Updatable(key="key2", value=input)


def llm_chain2(input: str):
    Updatable(key="key3", value=input)
    Updatable(key="key4", value=input)


def llm_chain(input: str):
    llm_chain1(input)
    llm_chain2(input)
```

```python
from IPython.display import display
from fastrepl import Analyze

with Analyze() as controller:
    llm_chain("")

    display(controller.build_graph(GRAPH=3))
```
