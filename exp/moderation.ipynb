```python
import openai

from fastrepl.polish import Updatable


def openai_1():
    input = Updatable(key="input", value="Sample text goes here")
    response = openai.Moderation.create(input=str(input))
    return input, response["results"][0]


def openai_2():
    input = Updatable(key="input", value="Sample text goes here")
    response = openai.Moderation.create(input=str(input))
    return input, response["results"][0]


def openai_3():
    openai_1()
    openai_2()
```

```python
from IPython.display import display
from fastrepl.loop import Analyze

with Analyze() as ctr:
    openai_3()

    display(ctr.build_graph(GRAPH=3))
```
