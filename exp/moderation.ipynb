```python
from fastrepl.utils import getenv
from IPython.display import display

import openai

openai.api_key = getenv("OPENAI_API_KEY", "")
```

```python
from fastrepl.polish import Updatable


def openai_1():
    input = Updatable(key="input", value="Sample text goes here")
    response = openai.Moderation.create(input=str(input))
    return input, response["results"][0]


def openai_2():
    input = Updatable(key="input", value="Sample text goes here")
    response = openai.Moderation.create(input=str(input))
    return input, response["results"][0]


def openai_3():
    openai_1()
    openai_2()
```

```python
from fastrepl.context import GlobalContext

GlobalContext.reset()
```

```python
from fastrepl.helpers import Experiment

with Experiment(GRAPH=3) as manager:
    openai_3()

    display(manager.build_graph())
```
